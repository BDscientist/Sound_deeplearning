```python

# 정위상

y, sr = librosa.load('/home/libedev/mute/mute-hero/download/dataset/sample_sound.wav', sr=16000)
time = np.linspace(0, len(y)/sr, len(y)) # time axis

fig, ax1 = plt.subplots() # plot
ax1.plot(time, y, color = 'b', label='speech waveform')
ax1.set_ylabel("Amplitude") # y 축
ax1.set_xlabel("Time [s]") # x 축
plt.title("정위상 파일 ") # 제목
plt.figure(1)


# 역위상

y1, sr1 = librosa.load('/home/libedev/mute/mute-hero/download/dataset/invert_sound2.wav', sr=16000)
time1 = np.linspace(0, len(y1)/sr1, len(y1)) # time axis

fig2, ax2 = plt.subplots() # plot
ax2.plot(time1, y1, color = 'b', label='speech waveform2')
ax2.set_ylabel("Amplitude") # y 축
ax2.set_xlabel("Time [s]") # x 축
plt.title("정위상 파일 ") # 제목
plt.figure(2)


plt.show()
```
