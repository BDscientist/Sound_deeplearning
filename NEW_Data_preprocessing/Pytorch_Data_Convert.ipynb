```python
import torch
import torch.nn as nn
import torch.optim as optim

import torchvision
import torchvision.utils as utils
import torchvision.datasets as dsets
import torchvision.transforms as transforms

import numpy as np
import random
import os
from torch.utils.data import Dataset, DataLoader




class my_datset(Dataset):
    
    
    
    
    
    def __init__(self, root, option, train = True):
        
        # root -->  new_train , new_test
        self.root = '/home/libedev/mute/mute-hero/download/dataset/'+root+'/'
        
        # option --> 전처리 폴더명
        self.option = option+'/'
        
        
        #add = self.root + self.option
        
        all_files = os.path.join(self.root,self.option)
        
        data=[os.path.join(dirpath,f)for dirpath, dirnames, filenames in os.walk(all_files) for f in filenames] 
        
        if self.train :
            
            
            
            self.train_data =[]
            self.train_label =[]
            self.train_label = self.train_label.append(np.load('/home/libedev/mute/mute-hero/download/dataset/new_train_label.npy'))
            
            
            for f in data:
                
                self.train_data.append(np.load(f))
                
                
            
            self.train_data = self.train_data.reshape((self.train_data.shape[0],1,self.train_data.shape[1],self.train_data.shape[2]))
            self.train_data = self.train_data.transpose((0,2,3,1))
            
            
            #완성된 dataset , label
            self.train_data = torch.from_numpy(self.train_data)   
            self.train_label = torch.from_numpy(self.train_label)

                
        else :
            
            
            self.test_data =[]
            self.test_label =[]
            
            self.test_label = self.test_label.append(np.load('/home/libedev/mute/mute-hero/download/dataset/new_test_label.npy'))
            
            
            
            for f in data:
                
                self.test_data.append(np.load(f))

            
            self.test_data=self.test_data.reshape(self.test_data.shape[0],1,self.test_data.shape[1],self.test_data.shape[2])
            self.test_data = self.test_data.transpose((0,2,3,1))
            
            
            #완성된 dataset , label
            self.test_data = torch.from_numpy(self.test_data)
            self.test_label = torch.from_numpy(self.test_label)
                
                
                
                
    
    def __len__(self , index):
        
            
        if self.train:
            return len(self.train_data)
            
        
        else:
            return len(self.test_data)
            

    
    
    def __getitem__(self,idx):
            
        if self.train :
            
            data , target = self.train_data[index], self.train_label[index]
            
        else:
                
            data, target = self.test_data[index], self.test_label[index]
                
        return data, target
    
    
    
```
